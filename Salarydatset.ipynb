```python
import matplotlib.pyplot as plt
plt.scatter(data['YearsExperience'], data['Salary'], color='green')
plt.title("Years of Experience vs salary")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x=data[['YearsExperience']]
y=data['Salary']
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=42)
model=LinearRegression()
model.fit(x_train, y_train)
print("Intercept", model.intercept_)
print("coefficient", model.coef_[0])
```

```python
n=int(input("Enter the number of years of experience"))
year=[[n]]
predicted_salary= model.predict (year)
print(f"Predicted salary for {n} years of experience:{predicted_salary[0]:.2f}")
```

```python
plt.scatter(data['YearsExperience'], data['Salary'], color='blue')
plt.plot(data['YearsExperience'], model.predict(data[['YearsExperience']]),color='red')
plt.title("after prediction: years of experience vs salary")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```
