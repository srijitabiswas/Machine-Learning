```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
def logistic(x,a,b,c):
  return a/(1+np.exp(-b*(x-c)))
```

```python
data=pd.read_csv('china_gdp.csv')
x1=data["Year"]
y=data["Value"]
p0=[max(y),0.02,1990]
```

```python
params,_=curve_fit(logistic,x1,y,p0)
a,b,c=params
print(f"Fitted parameter: a={a:.2f}, b={b:.4f}, c={c:.2f}")
```

```python
y_pred=logistic(x1,a,b,c)
```

```python
print("r2 score", r2_score(y,y_pred))
```

```python
plt.scatter(data['Year'], data['Value'], color='blue')
plt.plot(data['Year'], data['Value'], color='blue')
plt.title("China GDP")
plt.xlabel("Year")
plt.ylabel("Value")
plt.show()
```

```python
year=int(input("Enter the year: "))
predicted_gdp=logistic(year,a,b,c)
print("predicted gdp: ", predicted_gdp)
```
